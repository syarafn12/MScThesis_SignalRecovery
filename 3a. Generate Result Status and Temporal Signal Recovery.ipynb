```python
import rasterio
from rasterio.transform import from_origin
from pyproj import Proj, Transformer

# Open the TIFF file
with rasterio.open('Merdey_ASCENDING_VV_RQ2.tif') as src:
    tiff_data = src.read()
    transform = src.transform
    meta = src.meta

csv_data = mapping_rq1_vv_as # change data frame -> Descending VV (mapping_rq1_vv) or Ascending VV (mapping_rq1_vv_as) or Descending VH (mapping_rq1_vh) or Ascending VH (mapping_rq1_vh_as)

# Create an array for the new band, initialized with 2s for unmatched pixels
new_band = np.full((src.height, src.width), 9999, dtype=rasterio.float32)

# Transformer to convert from WGS 84 (latitude, longitude) to UTM zone 53S (EPSG:32753)
transformer = Transformer.from_crs("epsg:4326", "epsg:32753")

# Function to convert lat, lon to UTM and then to pixel coordinates
def latlon_to_pixel(lat, lon, transform):
    easting, northing = transformer.transform(lat, lon)
    return ~transform * (easting, northing)

# Loop through the CSV data
for index, row in csv_data.iterrows():
    pixel_coords = latlon_to_pixel(row['latitude'], row['longitude'], transform)
    pixel_x, pixel_y = map(int, pixel_coords)

    # Check if pixel coordinates are within the image bounds
    if 0 <= pixel_x < src.width and 0 <= pixel_y < src.height:
        new_band[pixel_y, pixel_x] = row['Duration100']
    else:
        print(f"Coordinates out of bounds for row {index}: {row['latitude']}, {row['longitude']}")

# Update metadata for the new band
meta.update(count = src.count + 1)

# Save the updated TIFF image with the new band
with rasterio.open('test.tif', 'w', **meta) as dst: # change new tiff image
    # Write existing bands
    for i in range(src.count):
        dst.write(tiff_data[i], i+1)
    # Write the new band
    dst.write(new_band, src.count + 1)
```
